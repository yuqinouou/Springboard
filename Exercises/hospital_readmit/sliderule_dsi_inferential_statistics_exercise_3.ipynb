# Hospital Readmissions Data Analysis and Recommendations for Reduction

### Background
In October 2012, the US government's Center for Medicare and Medicaid Services (CMS) began reducing Medicare payments for Inpatient Prospective Payment System hospitals with excess readmissions. Excess readmissions are measured by a ratio, by dividing a hospital’s number of “predicted” 30-day readmissions for heart attack, heart failure, and pneumonia by the number that would be “expected,” based on an average hospital with similar patients. A ratio greater than 1 indicates excess readmissions.

```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import bokeh.plotting as bkp
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
# read in readmissions data provided
hospital_read_df = pd.read_csv('data/cms_hospital_readmissions.csv')
```

## Preliminary Analysis

```python
# deal with missing and inconvenient portions of data 
clean_hospital_read_df = hospital_read_df[hospital_read_df['Number of Discharges'] != 'Not Available']
clean_hospital_read_df.loc[:, 'Number of Discharges'] = clean_hospital_read_df['Number of Discharges'].astype(int)
clean_hospital_read_df = clean_hospital_read_df.sort_values('Number of Discharges')
```

```python
# generate a scatterplot for number of discharges vs. excess rate of readmissions
# lists work better with matplotlib scatterplot function
x = [a for a in clean_hospital_read_df['Number of Discharges'][81:-3]]
y = list(clean_hospital_read_df['Excess Readmission Ratio'][81:-3])

fig, ax = plt.subplots(figsize=(8,5))
ax.scatter(x, y,alpha=0.2)

ax.fill_between([0,350], 1.15, 2, facecolor='red', alpha = .15, interpolate=True)
ax.fill_between([800,2500], .5, .95, facecolor='green', alpha = .15, interpolate=True)

ax.set_xlim([0, max(x)])
ax.set_xlabel('Number of discharges', fontsize=12)
ax.set_ylabel('Excess rate of readmissions', fontsize=12)
ax.set_title('Scatterplot of number of discharges vs. excess rate of readmissions', fontsize=14)

ax.grid(True)
fig.tight_layout()
```

The main issue for this analysis is that it was not conducted for different re-admission measures separately. The following plot displays how the expected readmission rate differs by each different type of readmissions. 

```python
# perform data cleaning
clean_hospital_read_df = clean_hospital_read_df[clean_hospital_read_df['Number of Discharges'] > 0]
```

```python
clean_hospital_read_df['Measure Name'].value_counts()
```

```python
clean_hospital_read_df.hist(column = 'Expected Readmission Rate', by = 'Measure Name', bins = 20,
                            sharex=True, figsize = (15, 3), layout = (1, 5))
plt.show()
```

Next we can check Excess Readmission Ratio (ERR) against Number of Discharge suggested by preliminary analysis.

```python
groups = clean_hospital_read_df.groupby('Measure Name')
plt.subplots(figsize = (16, 3))
for i, item in enumerate(groups):
    name, group = item
    plt.subplot(1, 5, i + 1)
    plt.margins(0.05)
    plt.plot(np.log10(group['Number of Discharges']), group['Excess Readmission Ratio'], 
                marker='.', linestyle='', ms=3, alpha = 1)
    plt.xlabel('Log(10) of Number of Discharge')
    plt.title(name)

plt.show()
```

When we look at the plots, we can see a pattern from hip and knee surgery panel that small hospitals seem to have higher ERR than big hospitals. To confirm that we run a quick hypothesis test. 

1. For hip and knee 30 days readmission  
H0: mean ERR for small hospitals(number of discharge <= 500) is the same as mean ERR for large hospitals.  
H1: mean ERR for small hospitals is different from mean ERR for large hospitals.  

```python
import scipy.stats as stats
hip_knee_read_df = clean_hospital_read_df[clean_hospital_read_df['Measure Name'] == 'READM-30-HIP-KNEE-HRRP']
hip_knee_read_df['large_hosp'] = hip_knee_read_df['Number of Discharges'] > 500
stats.ttest_ind(hip_knee_read_df[hip_knee_read_df['large_hosp']]['Excess Readmission Ratio'],
                hip_knee_read_df[~hip_knee_read_df['large_hosp']]['Excess Readmission Ratio'])
```

```python
hip_knee_read_df.groupby('large_hosp')['Excess Readmission Ratio'].median()
```

2./3. With p <0.01, the Excess Readmission Ratio of small hostpitals is significantly larger than Excess Readmission Ratio of Large hospitals, by around 0.11.

Statistical significance differs from practical significance in a way that even a statistical test signals, it does not necessarily means the result is interesting or concerning, if the magnitude of the association is too small to be noticed. For example, the hypothesis testing I ran above might also signal for another outcome, such as Pnemonia, but we might get a very small difference and not clinically meaningful. On the other hand, we might be more caucious about the outcomes such as AMI or HF since both are super sever clinical conditions associated with high death rates, and would like to report any signal we noticed.

5.  Advantages and Disadvantages for plots.   

Advantages:  
+ Fast - Usually can be generated within seconds.
+ Effective - Looking at hundreds of data points at once. 
+ Model-free - with small or none assumptions, easy to find hidden pattern of your data, very helpful for exploratory analysis.  
+ Visually-pleasant - easy to digest by audiance, if properly set up

Disadvantages:  
+ Lower-Dimensional - Cannot handle high dimensional data, start stuggling even starting from 3D.
+ Tunnel View - Can be mis-leading, easy to present marginal effect, not conditional. 
+ Cannot make statistical inference
